```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Invoices.csv")
```

```python
# Convert 'Date' and 'Date of Meal' to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Date of Meal'] = pd.to_datetime(df['Date of Meal'])
```

```python
df
```

```python
# Total Sales
total_sales = df['Meal Price'].sum()
print(f"Total Sales: ${total_sales}")
df
```

```python
# Sales by Type of Meal
sales_by_meal_type = df.groupby('Type of Meal')['Meal Price'].sum().sort_values(ascending=False)
print("\nSales by Type of Meal:\n", sales_by_meal_type)
```

```python
# Clean the Participants column by removing brackets and quotes
df['Participants'] = df['Participants'].str.strip("[]").str.replace("'", "")
```

```python
# Split participants into separate rows
participants_expanded = df['Participants'].str.split(', ', expand=True).stack()
```

```python
# Count the occurrences of each participant
participant_counts = participants_expanded.value_counts()
```

```python
# Get the top 5 most frequent participants
top_5_participants = participant_counts.head(5)
print("\nTop 5 Most Frequent Participants:\n", top_5_participants)
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales_trend = df.groupby('Month')['Meal Price'].sum()
```

```python
df
```

```python
# Plot Monthly Sales Trend
plt.figure(figsize=(10, 6))
monthly_sales_trend.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import seaborn as sns
```

```python
# Sales by Type of Meal
sales_by_meal_type = df.groupby('Type of Meal')['Meal Price'].sum().sort_values()

# Plot
plt.figure(figsize=(8, 4))
sns.barplot(x=sales_by_meal_type.index, y=sales_by_meal_type.values, palette="viridis")
plt.title('Total Sales by Type of Meal')
plt.xlabel('Type of Meal')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Distribution of Meal Types
meal_type_distribution = df['Type of Meal'].value_counts()

# Plot
plt.figure(figsize=(8, 8))
meal_type_distribution.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Meal Types')
plt.ylabel('')
plt.show()
```

```python
# Top 5 Most Frequent Participants (from previous processing)
top_5_participants = participants_expanded.value_counts().head(5)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_participants.values, y=top_5_participants.index, palette="coolwarm")
plt.title('Top 5 Most Frequent Participants')
plt.xlabel('Number of Meals')
plt.ylabel('Participants')
plt.show()
```

```python
# boxPlot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type of Meal', y='Meal Price', data=df, palette="Set3")
plt.title('Box Plot of Meal Prices by Type of Meal')
plt.xlabel('Type of Meal')
plt.ylabel('Meal Price')
plt.show()
```

```python
# histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Meal Price'], bins=30, kde=True, color='purple')
plt.title('Distribution of Meal Prices')
plt.xlabel('Meal Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatterPlot
plt.figure(figsize=(12, 6))
plt.scatter(df['Date of Meal'], df['Meal Price'], alpha=0.5, color='lightblue')
plt.title('Scatter Plot of Meal Prices Over Time')
plt.xlabel('Date of Meal')
plt.ylabel('Meal Price')
plt.show()
```
